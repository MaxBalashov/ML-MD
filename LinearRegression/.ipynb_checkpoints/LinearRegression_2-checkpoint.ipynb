# 2. Data
## 2.1. Data Import

```python
data = pd.read_csv('../data/prices.txt')
```

```python
data.shape
data.head()
```

```python
X = data[['area', 'rooms']].values
y = data['price'].values
```

## 2.2. Visualisation

```python
xlabel_lst = ('Area', 'Rooms')
```

```python
plt.figure(figsize=(7, 5))
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title('Scatter plot: Price ~ %s and %s' % xlabel_lst)
plt.ylabel('Area')
plt.xlabel('Rooms')
plt.grid();
```

```python
plt.figure(figsize=(15, 5))
for i in range(2):
    plt.subplot(1, 2, i + 1)
    plt.plot(X[:,i], y, 'o')
    plt.title('Price and %s' % xlabel_lst[i])
    plt.ylabel('Price')
    plt.xlabel(xlabel_lst[i])
    plt.grid()
    
plt.tight_layout();
```

# 3. Gradient Descent

```python
from LinearRegression_2 import LinearRegression
from LinearRegression_2 import loss, mae
from lr_utils import min_max_vec, linspace_vec
```

```python
LR = LinearRegression(alpha=0.0000001, n_iter=1000, fit_intercept=True, batch_size=10, random_state=42)
```

```python
LR.fit(X, y)
```

```python
y_pred = LR.predict(X)
```

```python
mae(y, y_pred)
loss(y, y_pred)
```

```python
# print('Number of epochs = %d' % LR.n_epoch)
print('R^2 = %.4f' % LR.score(X, y))

LR.w
```

```python
# Точки для построения графиков
X_min_max = min_max_vec(X)
X_linspase = linspace_vec(X_min_max)

coo = np.c_[np.ones(X_linspase.shape[1]), X_linspase.T]
y_pred = np.dot(coo, LR.w)
```

```python
plt.figure(figsize=(15, 5))
for i in range(1, 3):
    plt.subplot(1, 2, i)
    plt.plot(X[:, i - 1], y, 'o')
    plt.plot(coo[:, i], y_pred, '-')
    plt.title('Price and %s' % xlabel_lst[i - 1])
    plt.ylabel('Price')
    plt.xlabel(xlabel_lst[i - 1])
    plt.grid()
    
plt.tight_layout();
```
